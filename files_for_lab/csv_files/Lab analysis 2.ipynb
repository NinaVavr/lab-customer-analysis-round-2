```python
def change_to_lowercase(df):
    df.columns=[e.lower() for e in df.columns]
    return df
```

```python
data1 = change_to_lowercase(file1)
data1
```

```python
def reform_file(df):
    df=df.drop(columns=['unnamed: 0','vehicle type'])
    df=df.dropna()
    df['effective to date']=pd.to_datetime(df['effective to date'], errors='coerce')
    df['month'] = df['effective to date'].dt.month
    return df
```

```python
data2 = reform_file(data1)
data2
```

```python
# Split X- y
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler

X = file2.drop('total claim amount', axis=1)
y = file2['total claim amount']
X_num = X.select_dtypes(include =np.number)
X_cat = X.select_dtypes(exclude='number')
```

```python
X
```

```python
# Normalizing data: make data range from 0 - 1, instead of from min to max
transformer = MinMaxScaler().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
x_normalized
x_normalized2 =pd.DataFrame(x_normalized, columns=X_num.columns)
x_normalized2
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X_cat = X_cat.drop(['customer', 'effective to date'],  axis=1)
```

```python
encoder = OneHotEncoder(drop='first').fit(X_cat)
cols = encoder.get_feature_names_out(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(), columns=cols)
X_cat_encode.head()
```

```python
X_cat_encode.shape
```

```python
X = pd.concat([x_normalized2, X_cat_encode], axis=1 )
display(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```
